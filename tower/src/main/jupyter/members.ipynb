```python
import pandas as pd

# Read CSV into DataFrame
# Output from Mailchimp
data = pd.read_csv('subscriber/members.csv')
```

```python
# list the columns
# data.columns.tolist()
# The types
data.dtypes
```

```python
data['OPTIN_TIME'] = pd.to_datetime(data['OPTIN_TIME'])
data['day'] = pd.to_datetime(data['OPTIN_TIME'].dt.date)
data['month'] = pd.to_datetime(data['OPTIN_TIME'].dt.year.astype(str)+"-"+data['OPTIN_TIME'].dt.month.astype(str)+"-01")
# Check it with data[['Email Address','month']]

count_by_day = data.groupby('day').size().reset_index(name='count')
end_date = pd.to_datetime('2019-10-28')
count_by_day = count_by_day[count_by_day['day'] <= end_date] # filter
```

```python
import altair as alt
# Create Altair chart
chart = alt.Chart(count_by_day).mark_line().encode(
        x='day:T',  # 'T' tells Altair it's a temporal field
        y='count:Q',  # 'Q' for quantitative field
        tooltip=['day:T', 'count:Q']  # Define tooltip content
    ).properties(
        width=600,
        height=300,
        title='Subscription by Day'
    )
chart
```

```python
end_date_without_bot = pd.to_datetime('2019-10-17')
data_without_bot = data[data['day'] >= end_date_without_bot]
count_by_month = data_without_bot.groupby('month').size().reset_index(name='count')
count_by_month
```

```python

chart_month = alt.Chart(data_without_bot).mark_line().encode(
        x='month:T',
        y='count:Q',  # 'Q' for quantitative field
        tooltip=['day:T', 'count:Q']  # Define tooltip content
).properties(
        width=600,
        height=300,
        title='Subscription by Month'
    )
chart_month
```
